```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
max_score = 0
for score in employees.values():
    if score > max_score:
        max_score = score
top_performers = []
for name in employees:
    if employees[name] == max_score:
        top_performers.append(name)
names_str = ",".join(top_performers)
print(f"Top performers eligible for bonus: {names_str}(Score: {max_score})")
```

```python
sentence = "buy mobile phone buy phone online"
sentence = sentence.lower()
words = sentence.split()
count = {}
for word in words:
    if word in count:
        count[word] += 1
    else :
        count[word] = 1

result = {}
for word in count:
    if count[word] > 1:
        result[word] = count[word]
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] %2 == 0:
        valid_readings.append((i, sensor_readings[i]))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
g_count = 0
y_count = 0
for i in emails:
    if "@gmail.com" in i:
        g_count += 1
    elif "yahoo.com" in i:
        y_count += 1

total = len(emails)
print("gmail.com:",int((g_count /total) *100 ),"%")
print("yahoo.com:", int((y_count /total) *100 ),"%")  
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for i in sales:
    total += i
avg = total/len(sales)
limit = avg +(avg*0.20)
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i+1, ":", sales[i])
```
