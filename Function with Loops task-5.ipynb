```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")
sample_likes = [200, 350, 400, 100]
analyze_engagement(sample_likes)
```

```python
def analyze_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    for rainfall in rainfall_data:
        total_rainfall += rainfall
        average_rainfall = total_rainfall / len(rainfall_data)
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")
    return average_rainfall, status
sample_data =[60,80,75,73]
required_level_thresholf = 70
analyze_rainfall(sample_data, required_level_thresholf)
```

```python
def detect_duplicates(usernames):
    unique_usernames = set(usernames)
    if len(usernames) != len(unique_usernames):
        return "Duplicate Accounts Found : Yes"
    else:
        return "Accounts are unique"
user_list = ["user1", "user2", "user1", "user3"]
print(detect_duplicates(user_list))
```

```python
def check_app_health(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"
error_count = 7
ststus = check_app_health(error_count)
print(f"Error count: {error_count}")
print(f"System ststus: {ststus}")
```

```python
def check_transaction_limit(amount):
    daily_limit = 50000
    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"
transaction_amount = 60000
transction_status = check_transaction_limit(transaction_amount)
print(f"Transcation Amount: {transaction_amount}")
print(f"Transaction Status: {transction_status}")
```

```python
def analyze_bill(units):
    if units <= 100:
        bill = unit * 3
    elif units <= 200:
        bill = (100*3) +(units - 100)*5
    else:
        bill = (100*3) + (100*5) +(units - 200) *7
    if bill < 500:
        usage = "Low useage"
    elif bill <= 500:
        usage = "Moderate useage"
    else:
        usage = "High useage"
    return bill, usage
units = 250
amount, level = analyze_bill(units)
print(f"Bill: {amount}, Category: {level}")
```
