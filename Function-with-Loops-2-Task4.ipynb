```python
# Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Check minimum amount
        if amount < 50:
            print("❌ Invalid: Minimum recharge amount is ₹50")
            continue

        # Check valid plan
        if amount in valid_plans:
            print("✅ Recharge Successful!")
            break
        else:
            print("❌ Invalid plan. Valid plans are:", valid_plans)
            print("Please try again.\n")

# Call function
validate_recharge()
```

```python
# Inventory Reorder Alert System

def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"⚠️ Reorder Alert: {product} stock is low ({quantity})")
        else:
            print(f"✅ Stock OK: {product} ({quantity})")

# Example inventory
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

# Call function
check_inventory(inventory)
```

```python
# Student Result Processing System

def student_result(marks):
    total = 0

    # Loop to calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "✅ Pass"
    else:
        return "❌ Fail"

# Example marks
marks_list = [60, 55, 70, 40, 65]

# Call function
result = student_result(marks_list)
print("Result:", result)
```

```python
# Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        fare *= 1.25   # 25% extra

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print(f"Total Fare: ₹{total_fare:.2f}")

        retry = input("Retry? (yes/no): ")
        if retry.lower() != "yes":
            break

    except:
        print("Invalid input. Try again.")
```

```python
# Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_count = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage, "%")

    if percentage >= 75:
        return "✅ Eligible"
    else:
        return "❌ Not Eligible"


# Example attendance
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_attendance(attendance)
print("Eligibility:", result)
```

```python
# Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check length
    if len(password) < 8:
        return "❌ Weak Password (Minimum length 8 required)"

    # Loop through password
    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "✅ Strong Password"
    else:
        return "❌ Weak Password (Must include digit and special character)"


# Test password
password = input("Enter password: ")

result = check_password(password)
print(result)
```
