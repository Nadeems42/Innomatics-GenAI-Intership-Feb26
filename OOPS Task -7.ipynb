```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(self.name, "is", self.status)

light = SmartLight("Bedroom Light")
action = "ON"

if action == "ON":
    light.turn_on()
else:
    light.turn_off()

light.show_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

emp = Employee("Rahul", "EMP102", "IT")

emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

contact = Contact("Anita", "9876543210")

contact.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

p = Product("Headphones", 2499)

p.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

m = Movie("Inception", 4.8)

m.display()
```

```python
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)

d = Delivery("Suman", "Hyderabad")

d.display()
```
