```python
employees = {
    "Ravi":75000,
    "Anita":68000,
    "Kiran":72000
}
max_salary=0
max_employee=""
for name,salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        max_employee = name
print("Highest salary",max_employee,"-", max_salary)
```

```python
numbers = [45, 22, 89, 10, 66]
max_num = numbers[0]
min_num = numbers[0]
for i in numbers:
    if i > max_num:
        max_num = i
    elif i < min_num:
        min_num = i
print("List:", numbers)
print("Max:",max_num)
print("Min:",min_num)
```

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
    if i > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
count = 0
total = len(attendance)
for i in attendance:
    if i == "P":
        count += 1
        per = (count / total)*100
print("Attendence Percentage:",per)
```
